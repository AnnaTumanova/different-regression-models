```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"Stock_Price.csv")
indepX=df[["Interest_Rate","Unemployment_Rate"]]
depY=df["Stock_Index_Price"]
```

```python
#Generating Model Using Sklearn
from sklearn import linear_model
MLR_1=linear_model.LinearRegression()
MLR_1.fit(indepX,depY)
print("intercept:",MLR_1.intercept_)
print("coef:",MLR_1.coef_)
```
